#Airbnb New York City Listings  (Aluguéis de curta duração em NYC)

```python
# IMPORTAÇÃO DE BIBLIOTECAS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **1. Definição do Problema:**
Qual a questão a ser resolvida? O que esperamos descobrir com a análise?

1. Qual é a relação entre o preço das acomodações e o tipo de quarto?

2. Quais bairros têm as acomodações mais caras em média?

3. Como o número de avaliações varia entre os diferentes tipos de quarto?









#Significado das colunas

*   id: ID único de cada anúncio.
* name: Nome do anúncio ou título do imóvel.
* host_id: ID do anfitrião (proprietário ou gerente do imóvel).
* host_name: Nome do anfitrião.
* neighbourhood_group: Grupo de vizinhança principal em que o imóvel está
* localizado (ex: Brooklyn, Manhattan, etc.).
* neighbourhood: Nome do bairro específico dentro do grupo de vizinhança.
* latitude: Latitude da localização do imóvel.
* longitude: Longitude da localização do imóvel.
* room_type: Tipo de quarto ou imóvel, como:
* Entire home/apt: Imóvel inteiro.
* Private room: Quarto privado.
* Shared room: Quarto compartilhado.
* price: Preço por noite (em dólares) para a estadia no imóvel.
* minimum_nights: Número mínimo de noites exigidas para reserva.
* number_of_reviews: Total de avaliações feitas pelos hóspedes no anúncio.
* last_review: Data da última avaliação recebida.
* reviews_per_month: Média de avaliações recebidas por mês.
* calculated_host_listings_count: Número total de listagens ativas que o
* anfitrião possui.
* availability_365: Número de dias em que o imóvel está disponível para reserva durante o ano.

#2.Coleta de dados


```python
df = pd.read_csv("https://raw.githubusercontent.com/rillopedro/analisededadospython/refs/heads/main/AB_NYC_2019.csv")
df.head()
```

#Carregar e visualizar os dados

```python
# Verificando valores faltantes e nulos
df.isnull().sum()
```

```python
df.info()
```

```python
# Estatísticas descritivas
df.describe()
```

#Limpeza dos dados

```python
df = df.dropna()  # remove linhas com valores ausentes
df = df.drop_duplicates()  # remove duplicatas
```

```python
td = df.duplicated().sum()
print(f"Total de duplicatas encontradas: {td}")

df = df.drop_duplicates()

print(f"Total de duplicatas após a remoção: {df.duplicated().sum()}")
```

```python
td = df.dropna().sum()
print(f"Total de dados nulos encontrados: {td}")

df = df.dropna()

print(f"Total de dados nulos após a remoção: {df.dropna().sum()}")
```

#Visualização de dados


Perfunta 1: Qual é a relação entre o preço das acomodações e o tipo de quarto?

```python
# Filtrar preços para evitar outliers extremos que dificultem a visualização
filtro_preco = df[df['price'] < 500]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtro_preco, x='room_type', y='price')
plt.title('Distribuição de Preços por Tipo de Quarto')
plt.xlabel('Tipo de Quarto')
plt.ylabel('Preço (em dólares)')
plt.show()
```

Pergunta 2: Quais bairros têm as acomodações mais caras em média?

```python
media_vizinhanca = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(15)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=media_vizinhanca.values, y=media_vizinhanca.index, palette="viridis")
plt.title('Bairros com as Acomodações Mais Caras em Média')
plt.xlabel('Preço Médio (em dólares)')
plt.ylabel('Bairro')
plt.show()
```

Pergunta 3: Como o número de avaliações varia entre os diferentes tipos de quarto?

```python
total_de_avaliacoes = df.groupby('room_type')['number_of_reviews'].sum()
```

```python
plt.figure(figsize=(10, 6))
total_de_avaliacoes.plot(kind='bar', color='skyblue')
plt.title('Total de Avaliações por Tipo de Quarto')
plt.xlabel('Tipo de Quarto')
plt.ylabel('Total de Avaliações')
plt.xticks(rotation=0)  # Para manter os rótulos na horizontal
plt.show()
```

#Conclusão da Análise do Conjunto de Dados do Airbnb em Nova York
A análise do conjunto de dados do Airbnb em Nova York nos deu uma visão fascinante sobre como funcionam as acomodações na cidade. Aqui estão algumas das principais descobertas que emergiram dessa investigação:

1. Preços das Acomodações:

* Ao observar os preços por tipo de quarto, ficou claro que as acomodações inteiras costumam ter preços mais altos do que os quartos privados e compartilhados. Isso faz sentido, já que muitos viajantes valorizam a privacidade e a exclusividade de ter um espaço só para si, o que geralmente justifica o preço mais elevado.

2. Bairros Mais Caros:

* Quando analisamos os preços médios por bairro, notamos que algumas áreas, especialmente Manhattan, se destacam com os preços mais altos. Essas regiões são populares por sua localização central, atrações turísticas e uma variedade de serviços, tornando-se muito procuradas por turistas e moradores.

3. Distribuição de Avaliações:

* Ao explorar o número de avaliações, vimos que diferentes tipos de quarto atraem diferentes níveis de feedback. Os gráficos mostraram que os quartos privados e inteiros recebem mais avaliações, o que pode sugerir que esses tipos de acomodações proporcionam uma experiência mais satisfatória aos hóspedes ou que estão mais disponíveis.

4. Popularidade dos Tipos de Quarto:

* A análise da distribuição de avaliações revelou que a popularidade dos quartos varia. Os quartos privados, por exemplo, parecem ser um ótimo compromisso entre conforto e preço, atraindo um número maior de hóspedes. Isso pode explicar por que eles recebem mais feedback.
Considerações Finais
Essas descobertas são valiosas tanto para os anfitriões que buscam se destacar no mercado quanto para os viajantes que querem encontrar as melhores opções de aluguel. A diferença nos preços e avaliações destaca a importância de escolher bem o bairro e o tipo de acomodação ao alugar um imóvel pelo Airbnb em Nova York. Para os anfitriões, é fundamental acompanhar o mercado e as preferências dos hóspedes, ajustando preços e ofertas quando necessário para permanecerem competitivos.
